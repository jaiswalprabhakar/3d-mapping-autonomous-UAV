```python
import cv2
import numpy as np

frame = cv2.imread("image2.jpg")
# It converts the BGR color space of image to HSV color space
hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
# Threshold of blue in HSV space
lower_value = np.array([20, 80, 0])
upper_value = np.array([255, 255, 255])

# preparing the mask to overlay
mask = cv2.inRange(hsv, lower_value, upper_value)
    
# The black region in the mask has the value of 0,
# so when multiplied with original image removes all non-blue regions
image = cv2.bitwise_and(frame, frame, mask = mask)
cv2.imshow('filtered', image)
# Grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Find Canny edges
edged = cv2.Canny(gray, 10, 20)

###Morphology
se = np.ones((7,7), dtype='uint8')
image_close = cv2.morphologyEx(image, cv2.MORPH_CLOSE, se)
cv2.imshow('Morphology', image_close)

# Finding Contours
area = 0
contours, hierarchy = cv2.findContours(edged, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
for i in range(len(contours)):
    print(cv2.contourArea(contours[i]))
    area += cv2.contourArea(contours[i])


cv2.imshow('Canny Edges After Contouring', edged)

print("Number of Contours found = " + str(len(contours)))
print("COuntour Area= " + str(area))

# Draw all contours
# -1 signifies drawing all contours
cv2.drawContours(image, contours, -1, (0, 0, 255), 1)

cv2.imshow('Contours', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
